```python

import cv2
import matplotlib.pyplot as plt

# Turn off warnings (not directly equivalent, but similar effect)
import warnings
warnings.filterwarnings("ignore")
```

```python
x = cv2.imread('/content/dog.jpeg')
# Convert to RGB (matplotlib uses RGB)
x_rgb = cv2.cvtColor(x, cv2.COLOR_BGR2RGB)
```

```python
# Display the original image
plt.imshow(x_rgb)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# Convert to grayscale
x_gray = cv2.cvtColor(x, cv2.COLOR_BGR2GRAY)
```

```python
# Apply Canny edge detection
w = cv2.Canny(x_gray, 50, 150)
```

```python
# Display the edge-detected image
plt.imshow(w, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')
plt.show()
```

```python
# Calculate the average value of edges
sum_value = w.sum()
r, c = w.shape
ls = sum_value / (r * c)
```

```python
print("Average value of edges:", ls)
```
